```
pd.read_sql('''SELECT nov_01_tweets.user_id as user, user_active_status, COUNT(nov_01_tweets.tweet_id) as total_tweets 
            FROM nov_01_tweets 
            CROSS JOIN all_users ON nov_01_tweets.user_id = all_users.user_id 
            GROUP BY user 
            ORDER BY total_tweets
            DESC LIMIT 100''', conn)
```

```
curr.execute('''DROP VIEW IF EXISTS V_user_and_tweets''')
curr.execute('''CREATE VIEW V_user_and_tweets AS SELECT  
                nov_01_tweets.user_id as user,
                COUNT(nov_01_tweets.tweet_id) as total_tweets 
            FROM nov_01_tweets 
            GROUP BY user 
            ORDER BY total_tweets
            ''')
curr.execute('''DROP VIEW IF EXISTS V_user_status_tweets''')
curr.execute('''CREATE VIEW V_user_status_tweets AS SELECT V_user_and_tweets.user, V_user_and_tweets.total_tweets, all_users.user_active_status 
                FROM V_user_and_tweets 
                JOIN all_users ON V_user_and_tweets.user = all_users.user_id
                GROUP BY user
                ORDER BY total_tweets
                DESC''')
curr.execute('''DROP VIEW IF EXISTS V_user_status_tweets_re''')
curr.execute('''CREATE VIEW V_user_status_tweets_re AS SELECT 
                nov_01_retweets.user_id as user_id, COUNT(nov_01_retweets.tweet_id) as retweet_count , V_user_status_tweets.total_tweets as tweet_count, V_user_status_tweets.user_active_status as status
                FROM nov_01_retweets 
                JOIN V_user_status_tweets ON nov_01_retweets.user_id = V_user_status_tweets.user
                GROUP BY nov_01_retweets.user_id 
                ORDER BY retweet_count 
                DESC''')
pd.read_sql("SELECT * FROM V_user_status_tweets_re", conn) #users that tweeted on november 1, their status, and the total number of tweets they created on that date, and the total number of retweets they made
```
